```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
import datetime
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
from sklearn.preprocessing import StandardScaler, OneHotEncoder 
import matplotlib.pyplot as plt
```

### Data loading and processing 

```python
test=pd.read_csv(r"C:\Users\sophi\OneDrive\Bureau\ElevenCase\OneDrive_2019-09-18\testsetCleanFinal.csv", sep=",")
```

```python
test.head()
```

```python
ProfileReport(test)
```

```python
test.drop(labels=["WSF2max","WSF5max"], axis=1, inplace=True)
```

```python
test.ActualLandingTime=pd.to_datetime(test.ActualLandingTime.values)
test.Date=pd.to_datetime(test.Date.values)
```

```python
test.head()
```

```python
train=pd.read_csv(r"C:\Users\sophi\OneDrive\Bureau\ElevenCase\OneDrive_2019-09-18\DatasetTrainingFinal.csv", sep=",")
```

```python
train.drop(labels=["WSF2max","WSF5max"], axis=1, inplace=True)
train.ActualLandingTime=pd.to_datetime(train.ActualLandingTime.values)
train.Date=pd.to_datetime(train.Date.values)
```

### 1st model (10 variables)

```python
X_train=train.iloc[:,6:16]
y_train=train.iloc[:,22:23]
X_test=test.iloc[:,6:16]
y_test=test.iloc[:,22:23]
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
model=LinearRegression(normalize=False)

X_train=X_train.values.reshape(-1,10)
y_train=y_train.values.reshape(-1,1)
X_test=X_test.values.reshape(-1,10)
y_test=y_test.values.reshape(-1,1)

model.fit(X_train, y_train)
r_sq = model.score(X_train, y_train)

y_pred=model.predict(X_test)
y_accur=model.predict(X_train)

MAE=metrics.mean_absolute_error(y_test, y_pred)
RMSE_test=np.sqrt(metrics.mean_squared_error(y_test, y_pred))
RMSE_train=np.sqrt(metrics.mean_squared_error(y_train, y_accur))

print("R2_train :", r_sq)
print("RMSE_test :", RMSE_test)
print("RMSE_train :", RMSE_train)
print("MAE_test : ", MAE)
```

```python
model.coef_
```

#### Variables utilisÃ©es : 
AWNDmean, PRCPmean, TAVGmean, WDF5mean, WT01max, WT02max, WT03max, WT08max, IceRoad, WetDay

```python
ecart_test = y_pred-y_test

score_3min=((ecart_test>=-3) & (ecart_test<=3)).sum()/ecart_test.shape[0]
print("Accuracy_test at 3 min", round(score_3min,10))

score_5min=((ecart_test>=-5) & (ecart_test<=5)).sum()/ecart_test.shape[0]
print("Accuracy_test at 5 min", round(score_5min, 10))

ecart_train = y_accur-y_train

score_3min=((ecart_train>=-3) & (ecart_train<=3)).sum()/ecart_train.shape[0]
print("Accuracy_train at 3 min", round(score_3min,10))

score_5min=((ecart_train>=-5) & (ecart_train<=5)).sum()/ecart_train.shape[0]
print("Accuracy_train at 5 min", round(score_5min, 10))
```
